# 당뇨병 데이터

```python
import pandas as pd
```

```python
dt=pd.read_csv('./diabetes.csv')
```

```python
df=pd.DataFrame(dt)
df
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#maxdepth =[5,6,7]
```

```python
x = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
        'BMI', 'DiabetesPedigreeFunction', 'Age']]
y = df[['Outcome']] # 레이블 넣어주기
```

```python

X_train,X_test , y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2, random_state=121)

dtree = DecisionTreeClassifier()

parameters={'max_depth':[5,6,7],'min_samples_split':[2,3]} # 순서 여섯개 섞는
```

```python
grid_dtree =GridSearchCV(dtree,param_grid=parameters,cv=3,refit=True)

grid_dtree.fit(X_train,y_train) # 훈련

scores_df=pd.DataFrame(grid_dtree.cv_results_)
scores_df[['params','mean_test_score','rank_test_score','split0_test_score',
           'split1_test_score','split2_test_score']]
```

```python
print('GridSearchCV 최적 파라미터:', grid_dtree.best_params_)
print('GridSearchCV 최고 정확도: {0:4f}'.format(grid_dtree.best_score_))
```

```python
#GridSearchCV의 refit 으로 이미 학습이 된 etimator 반환
estimator=grid_dtree.best_estimator_
#GridSearchCV의 best_estimator_는 이미 최적 하이퍼 파라미터로 학습이 됨.
pred=estimator.predict(X_test)
print(f"테스트 데이터 세트 정확도 {accuracy_score(y_test, pred):.4f}")
```
